Getting geospatial data from CSV.

```python
filename = 'http://cocl.us/Geospatial_data'
df_coordinates = pd.read_csv(filename)
```

```python
df_coordinates
```

Merging Postal Codes with Coordinates

```python
df_new =  pd.merge(df,df_coordinates, on ='Postal Code', how ='left') 
df_new
```

```python
df_new.shape
```

Getting all Boroughs with the name "Toronto"

```python
def_tor = df_new[df_new['Borough'].str.contains('Toronto',regex=False)]
def_tor
```

```python
!conda install -c conda-forge folium=0.5.0 --yes
import folium
```

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(def_tor['Latitude'],def_tor['Longitude'],def_tor['Borough'],def_tor['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
k=5
toronto_clustering = def_tor.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
def_tor.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
def_tor
```

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(def_tor['Latitude'], def_tor['Longitude'], def_tor['Neighbourhood'], def_tor['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
